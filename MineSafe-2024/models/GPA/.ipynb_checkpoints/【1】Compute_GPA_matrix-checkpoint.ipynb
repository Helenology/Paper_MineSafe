```python
# Raw Image Failes
raw_files = glob.glob("../../dataset/baseline/highway/input/*jpg")
raw_files.sort()

# Example Raw Image
raw_img = Image.open(raw_files[0])
raw_img = np.array(raw_img).astype(np.int32) / 255.0
```

```python
G = 500 # number of grid points
p = raw_img.shape[0]
q = raw_img.shape[1]
N0 = len(raw_img) # number of images to compute the GPA matrix
```

```python
shape = (p, q)
# tick_tensor = tf.concat([tf.ones([1, p, q]) * tick for tick in tick_list], axis=0)
# print(tick_tensor.shape)
alpha = np.log(p * q) / np.log(N0)
bandwidth, bandwidth_star = compute_optimal_bandwidths(N0, alpha)
```

```python
rng = np.random.default_rng(seed=1)
# img_index = rng.integers(low=0, high=int(0.9 * N), size=N0, dtype='int64')
# print(len(img_index))

# with open('./img_index.pkl', "wb") as f:
#     pickle.dump(img_index, f)
```

```python
tick_list = rng.random(size=G)
with open('./tick_list.pkl', "wb") as f:
    pickle.dump(tick_list, f)
```

```python
tick_tensor = tf.concat([tf.ones([1, p, q]) * tick for tick in tick_list], axis=0)
tick_tensor.shape
```

```python
j = 0
GPA_matrix = tf.zeros([G, p, q])

for i in range(len(raw_files)):
    print(f"--Load the {j}th image to compute GPA matrix with [index={i}]--")
    test_img = load_and_preprocess_image(raw_files[i], shape)
    tmp_tensor = 1 / tf.sqrt(2*np.pi) * tf.exp(-(test_img - tick_tensor)**2 / (2 * bandwidth**2))
    GPA_matrix += tmp_tensor
    j += 1
```

```python
with open('./GPA_matrix.pkl', 'wb') as f:
    pickle.dump(GPA_matrix, f)
```

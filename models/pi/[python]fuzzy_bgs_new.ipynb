```python
def img_read(file_name):
    img_array = Image.open(file_name)
    img_array = np.array(img_array) / 255.0  # with shape (p, q, 3)
    return img_array   
```

```python
def mean_filter(im1, im2, mean_window):
    if mean_window == 1:
        im1 = im2 * 1.0
    else:
        aa = int((mean_window - 1) / 2)
        im1 = im2 * 0.0
        for ro in range(aa, M-aa):
            for cl in range(aa, N-aa):
                for dp in range(C):
                    for xo in range(ro-aa, ro+aa+1):
                        for yo in range(cl-aa, cl+aa+1):
                            im1[ro][cl][dp] = im1[ro][cl][dp] + im2[xo][yo][dp]
                    im1[ro][cl][dp] = im1[ro][cl][dp] / (mean_window * mean_window)
    return im1, im2
```

```python
def median_filter(im3, im6, med_window):
    ar = 0
    ro = 0
    cl = 0
    dp = 0
    m = 0
    
    if med_window == 1:
        im6 = im3 * 1.0
    else:
        aa = (med_window - 1) / 2
        bb = (((med_window * med_window) - 1) / 2) + 1
        ar = np.zeros(med_window * med_window)
        for dp in range(C):
            for ro in range(aa, M-aa):
                for cl in range(aa, N-aa):
                    m = 0
                    for i in range(ro-aa, ro+aa+1):
                        for j in range(cl-aa, cl+aa+1):
                            ar[m] = im3[i][j][dp]
                            m = m + 1
                    for i in range(med_window*med_window):
                        for j in range(med_window*med_window-1):
                            if ar[j] >= ar[j+1]:
                                uu = ar[j+1]
                                ar[j+1] = ar[j]
                                ar[j] = uu
                    im6[ro][cl][dp] = ar[bb]
    return im3, im6
```

```python
print("\n***********Background Training starts here*********\n");
for img_path in train_img_list:
    im2 = img_read(img_path)
    print(f"taking input file:\n{img_path}");
    im1, im2 = mean_filter(im1, im2, mean_window)
    print("**************************success********************************\n")
    temp3 = 0.0
#     temp4 = 0
    temp6 = 2.0 / (fuz - 1)
    temp7 = 0.0
    temp8 = 0.0

    for i in range(w, M-w):
        for j in range(w, N-w):
            for r in range(1, flag[i][j] + 1):
                for k in range(C):
                    mean_temp[i][j][r][k] = mean[i][j][r][k]
            for r in range(1, flag[i][j] + 1):
                dis_temp[i][j][r] = 0.0
            for r in range(1, flag[i][j] + 1):
                for k in range(C):
                    dis_temp[i][j][r] = dis_temp[i][j][r] + np.power(abs(mean_temp[i][j][r][k]-im1[i][j][k]), 2.0)
                dis_temp[i][j][r] = np.sqrt(dis_temp[i][j][r])
            for r in range(1, flag[i][j] + 1):
                mu_temp[i][j][r] = 0.0
            for r in range(1, flag[i][j] + 1):
                mu_temp[i][j][r] = 0.0
                if dis_temp[i][j][r] == 0.0:
                    for s in range(1, flag[i][j] + 1):
                        mu_temp[i][j][s] = 0.0
                    mu_temp[i][j][r] = 1.0
                    break
                else:
                    temp3 = 0.0
                    for s in range(1, flag[i][j] + 1):
                        temp3 = temp3 + np.power((dis_temp[i][j][r] / dis_temp[i][j][s]), temp6)
                    mu_temp[i][j][r] = 1 / temp3
            temp7 = cost[i][j]
            temp8 = cost[i][j]
            for r in range(1, flag[i][j] + 1):
                temp7 = temp7 + np.power(mu_temp[i][j][r], fuz) * np.power(dis_temp[i][j][r], 2)
            dis_temp[i][j][flag[i][j]+1] = 0.0
            for k in range(C):
                p = 0
                for m in range(i-w, i+w+1):
                    for n in range(j-w, j+w+1):
                        ar[k][p] = im1[m][n][k]
                        p = p + 1
            for k in range(C):
                temp2[i][j][k] = 0.0
                for p in range(W * W):
                    temp2[i][j][k] = temp2[i][j][k] + ar[k][p]
            for k in range(C):
                mean_temp[i][j][flag[i][j]+1][k] = temp2[i][j][k] / (W * W)
            for r in range(1, flag[i][j] + 1):
                for k in range(C):
                    mean_temp[i][j][r][k] = mean[i][j][r][k]
            for r in range(1, flag[i][j]+2):
                dis_temp[i][j][r] = 0.0
                for k in range(C):
                    dis_temp[i][j][r] = dis_temp[i][j][r] + np.power(abs(mean_temp[i][j][r][k]-im1[i][j][k]), 2.0)
                dis_temp[i][j][r] = np.sqrt(dis_temp[i][j][r])
            for r in range(1, flag[i][j]+2):
                mu_temp[i][j][r] = 0.0
            for r in range(1, flag[i][j]+2):
                mu_temp[i][j][r] = 0.0
                if dis_temp[i][j][r] == 0.0:
                    for s in range(1, flag[i][j]+2):
                        mu_temp[i][j][s] = 0.0
                    mu_temp[i][j][r] = 1.0
                    break
                else:
                    temp3 = 0.0
                    for s in range(1, flag[i][j]+2):
                        temp3 = temp3 + np.power((dis_temp[i][j][r] / dis_temp[i][j][s]), temp6)
                    mu_temp[i][j][r] = 1 / temp3
            for r in range(1, flag[i][j]+2):
                temp8 = temp8 + np.power(mu_temp[i][j][r], fuz) * np.power(dis_temp[i][j][r], 2)
            if (temp7 < temp8) or (abs(temp7 - temp8) < thr):
                for r in range(1, flag[i][j] + 1):
                    for k in range(C):
                        mean[i][j][r][k] = ((occur[i][j][r]*mean[i][j][r][k])+(mu_temp[i][j][r]*im1[i][j][k]))/(occur[i][j][r]+mu_temp[i][j][r])
                    occur[i][j][r] = occur[i][j][r] + mu_temp[i][j][r]
            else: 
                for r in range(1, flag[i][j]+1):
                    for k in range(C):
                        mean[i][j][r][k] = ((occur[i][j][r]*mean[i][j][r][k])+(mu_temp[i][j][r]*im1[i][j][k]))/(occur[i][j][r]+mu_temp[i][j][r])
                    occur[i][j][r] = occur[i][j][r]+mu_temp[i][j][r]

                for k in range(C):
                    mean[i][j][flag[i][j]+1][k] = mean_temp[i][j][flag[i][j]+1][k]
                occur[i][j][flag[i][j]+1] = mu_temp[i][j][flag[i][j]+1]
                flag[i][j] += 1

    for i in range(M):
        for j in range(N):
            for T in range(1, K):
#                 temp5[T] = 0.0
                for k in range(C):
                    temp1[i][j][T][k]=0.0
                    temp2[i][j][k]=0.0
    temp3=0
#                     temp4=0
    temp7=0.0
    temp8=0.0
```
